```python
import pandas as pd
```

```python
result = pd.read_csv('./data/result.csv')
cross = pd.crosstab(result['user_id'], result['aisle'])
cross.head(10)
```

```python
# 压缩稀疏矩阵，进行主成分分析
from sklearn.decomposition import PCA
pca = PCA(n_components = 0.9)
data = pca.fit_transform(cross)
```
